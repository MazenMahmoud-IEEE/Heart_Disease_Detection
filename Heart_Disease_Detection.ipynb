# Importing Libraries

```python
import pandas as pd
import numpy as np
from IPython.display import display
import seaborn as sns
import matplotlib.pylab as plt
import warnings
warnings.filterwarnings('ignore')
```

# Reading and copying the dataset

```python
data = pd.read_csv("heart_dataset.csv")
print(data.shape)
display(data)
data_cpy = data.copy()
```

# Removing Duplicates

```python
data_cpy = data_cpy.drop_duplicates()
display(data_cpy)
```

# Dealing with Nulls

```python
nulls = data_cpy.isnull().sum().sort_values(ascending=False)
print(nulls)
```

# Data Preprocessing

```python
data_cpy.info()
```

```python
data_cpy.info()
```

```python
data_cpy.select_dtypes(include='object').nunique()
```

```python
data_cpy['Sex'] = data_cpy['Sex'].str.strip().map({'Female': 0, 'Male': 1})
data_cpy
```

# Multicategory columns

```python
# List of columns with more than two unique values
multi_category_columns = ['AgeCategory','Race','Diabetic','GenHealth']

# Apply one-hot encoding
features = pd.get_dummies(data_cpy, columns=multi_category_columns, drop_first=True)  
display(features)
```

```python
features.drop('Race_Other' , axis = 1 , inplace = True)
features
```

# Boolean Columns

```python
binary_columns = [
    "HeartDisease",
    "Smoking",
    "AlcoholDrinking",
    "Stroke",
    "DiffWalking",
    "Sex",
    "PhysicalActivity",
    "Asthma",
    "KidneyDisease",
    "SkinCancer"
]

# Convert binary columns to 0 and 1
features[binary_columns] = features[binary_columns].replace({'Yes': 1, 'No': 0})
features
```

# Converting boolean columns (True & False) into (0 & 1)

```python
# Convert boolean values to 0 and 1
features = features.astype(float)

# Display the updated DataFrame
display(features)
```

```python
features.info()
```

# Balancing the data (Undersampling and Oversampling)

## Undersampling

```python
from imblearn.over_sampling import SMOTE
from imblearn.over_sampling import ADASYN

# Step 1: Take all instances from the minority class (HeartDisease == 1)
x_1 = features[features['HeartDisease'] == 1]
y_1 = features[features['HeartDisease'] == 1]
y_1 = y_1['HeartDisease']
x_1 = x_1.drop('HeartDisease' , axis = 1)

display(x_1)
print()
display(y_1)
print()

# Step 2: Sample 185,000 instances (70 % of the data ) from the majority class (HeartDisease == 0)
x_0 = features[features['HeartDisease'] == 0].sample(n=189000, random_state=0)
y_0 = features[features['HeartDisease'] == 0].sample(n=189000, random_state=0)
y_0 = y_0['HeartDisease']
x_0 = x_0.drop('HeartDisease' , axis = 1)

display(x_0)
print()
display(y_0)
print()
```

## Store dropped data to be tested later

```python
# Get the indices of the sampled data (x_0)
sampled_indices = x_0.index
indices = x_1.index

# Get the remaining (non-sampled) data by excluding those indices
test_x = features.drop(sampled_indices)
test_x = test_x.drop(indices)
test_x = test_x.reset_index()
display(test_x)

test_x.drop('index', axis = 1 , inplace = True)
test_y = test_x['HeartDisease']
test_x = test_x.drop('HeartDisease' , axis = 1)

display(test_x)
print()
display(test_y)
print()
```

## Resetting index for both datasets before merging

```python
x_0 = x_0.reset_index()
x_0.drop('index',axis=1,inplace=True)
x_0
```

```python
y_0 = y_0.reset_index()
y_0.drop('index',axis=1,inplace=True)
y_0
```

## Merging datasets

```python
X_balanced = pd.concat([x_0, x_1])
y_balanced = pd.concat([y_0, y_1])

display(X_balanced)
display(y_balanced)
```

# Oversampling

```python
from imblearn.over_sampling import SMOTE

# Create SMOTE object
smote = SMOTE(sampling_strategy={1: 63000}, random_state=0)
X_resampled, y_resampled = smote.fit_resample(X_balanced, y_balanced)

# Create a new DataFrame with the resampled data
features = pd.DataFrame(X_resampled, columns= X_resampled.columns)
features["HeartDisease"] = y_resampled

display(features)

# Nearly 55 % of the minority class and 14 % of the total data are artificial points 
# We had to do this due to the great class ratio of 10 : 1
```

# Correlation

```python
corr = features.corr()['HeartDisease']
corr
```

```python
top_features = corr[(corr > 0.1) | (corr < - 0.1)].sort_values(ascending=False) #displays  correlation coefficient greater than 0.5
top_features = top_features.drop('HeartDisease')
top_features
```

# Features and Target

```python
target = features['HeartDisease']
target
```

```python
features.drop('HeartDisease' ,axis=1 , inplace=True)
features = features[top_features.index]
display(features)
```

```python
test_x = test_x[top_features.index]
display(test_x)
```

# Splitting the data for training and testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=0)
```

```python
target.value_counts()
# There is class imbalance in the data
```

# Testing all base models

```python
# Importing necessary libraries
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
import xgboost as xgb
from sklearn.ensemble import AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.metrics import f1_score, accuracy_score
from sklearn.model_selection import ShuffleSplit
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import ADASYN
from imblearn.under_sampling import RandomUnderSampler
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler

# Function to build, train, and evaluate the models
def model_building(model_name):
    model = model_name
    model.fit(X_train, y_train)
    print(f"******** Model :- {model_name.__class__.__name__} ********")

    y_test_pred = model.predict(X_test)
    y_train_pred = model.predict(X_train)

    train_f1 = f1_score(y_train, y_train_pred)
    test_f1 = f1_score(y_test, y_test_pred)
    train_accuracy = accuracy_score(y_train, y_train_pred)
    test_accuracy = accuracy_score(y_test, y_test_pred)

    print(f"Training F1 Score: {train_f1:.4f}, Training Accuracy: {train_accuracy:.4f}")
    print(f"Testing F1 Score: {test_f1:.4f}, Testing Accuracy: {test_accuracy:.4f}")
    print()

# Dictionary with different models
model_dict = {
    'dt': DecisionTreeClassifier(random_state=0, class_weight='balanced'),
     #'knn': KNeighborsClassifier(),
    'rf': RandomForestClassifier(criterion='entropy' , random_state=0, class_weight='balanced'),
    'xgb': xgb.XGBClassifier(random_state=0, class_weight='balanced'),
    'ada': AdaBoostClassifier(random_state=0)
    #'dl' : MLPClassifier(random_state=0)
    #,'svm':SVC(random_state=0,class_weight='balanced')
}

# Calling to build and evaluate models
for key in model_dict.keys():
    model_building(model_dict[key])
```

## Found best base model : Gradient Boost

# Tuning Gradient boost Model

## BayesianSearch

```python
from skopt import BayesSearchCV
from xgboost import XGBClassifier


# Define the hyperparameter search space
param_space = {
    'learning_rate': (0.01, 0.5, 'log-uniform'),
    'max_depth': (2,6),
    'n_estimators': (100, 500),
    'subsample': (0.5, 1.0),
    'colsample_bytree': (0.56, 1),
    'min_samples_split': (2,4),
    'min_samples_leaf': (2,4)   
}

xgb_model = XGBClassifier(random_state = 0 , class_weight = 'balanced' )
bayes_search = BayesSearchCV(estimator=xgb_model, search_spaces=param_space, n_iter=30, cv=3 , scoring='f1')

# Fit the model to find the best hyperparameters
bayes_search.fit(X_train, y_train)
best_model = bayes_search.best_estimator_
y_train_pred = best_model.predict(X_train)
y_test_pred = best_model.predict(X_test)

# Calculate F1 score and accuracy for training and testing sets
train_f1 = f1_score(y_train, y_train_pred)
test_f1 = f1_score(y_test, y_test_pred)
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)

# Output the best parameters found
print(bayes_search.best_params_)
print(f"Training F1 Score: {train_f1:.4f}, Training Accuracy: {train_accuracy:.4f}")
print(f"Testing F1 Score: {test_f1:.4f}, Testing Accuracy: {test_accuracy:.4f}")
```

# Results

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix

cf_matrix = confusion_matrix(y_test, y_test_pred)
group_names = ['True Neg','False Pos','False Neg','True Pos']

group_counts = ["{0:0.0f}".format(value) for value in
                cf_matrix.flatten()]

group_percentages = ["{0:.2%}".format(value) for value in
                     cf_matrix.flatten()/np.sum(cf_matrix)]
labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in
          zip(group_names,group_counts,group_percentages)]

labels = np.asarray(labels).reshape(2,2)

sns.heatmap(cf_matrix, annot=labels, fmt='', cmap='Blues')
```

# Testing on random patients from the dropped non seen data

```python
# Function to check the output for a specific customer
def check_disease_by_index(patient_index, best_model, test_x, test_y=None):

        # Retrieve the customer's data by index
    patient_data = test_x.iloc[patient_index]

        # Reshape the data to be 2D for prediction
    patient_data = patient_data.values.reshape(1, -1)

        # Predict the outcome for the customer
    patient_data = best_model.predict(patient_data)

        # Output the result
    print(f"Prediction for Patient at index {patient_index}: {patient_data[0]}")

        # If true labels exist, compare the prediction to the actual label
    if y_test is not None:
        actual_output = test_y.iloc[patient_index]
        print(f"Actual Output for Patient at index {patient_index}: {actual_output}")

    return patient_data , actual_output


# Replace with the actual index of the customer you want to check
True_positives , True_negatives , False_positives , False_negatives = 0 , 0 , 0 , 0

for i in range(0,10000):
    patient_data = i * 3
    patient_prediction , actual_output = check_disease_by_index(patient_data, best_model, test_x, test_y)
    if (patient_prediction == actual_output) and (actual_output == 1) :
        True_positives = True_positives + 1
    elif (patient_prediction == actual_output) and (actual_output == 0):
        True_negatives = True_negatives + 1
    elif (patient_prediction != actual_output) and (actual_output == 0) :
        False_positives = False_positives + 1
    else :
        False_negatives = False_negatives + 1

print()
print("True positives = ",True_positives)
print("True negatives = ",True_negatives)
print("False positives = ",False_positives)
print("False negatives = ",False_negatives)
```
